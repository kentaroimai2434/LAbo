```python
import numpy as np
import pandas as pd
import os
from SimpleCycleReservoir import SimpleCycleReservoir
import matplotlib.pyplot as plt
import japanize_matplotlib
from sklearn.metrics import mean_squared_error as MSE
import random
from tqdm import tqdm
from multiprocessing import Pool,cpu_count
from signalz import mackey_glass
```

```python
data = pd.read_csv(os.path.join('Henon_Map_time_series_ex1.csv'))
train_inputs = data['values'][:1000]
train_target = data['values'][1:1001]
test_data = data['values'][1000:].reset_index(drop=True)
data.head()
```

```python
mg = mackey_glass(5200, a=0.2, b=1, c=0.9, d=17, e=10, initial=0.1)
plt.plot(mg[200:])
train_inputs = mg[200:1200]
train_target = mg[201:1201]
test_data = mg[1200:]
```

### モデル構築

```python
NUM_INPUT_NODES = 1
NUM_RESERVOIR_LAYER = 2
NUM_RESERVOIR_NODES = 75
NUM_OUTPUT_NODES = 1
NUM_EPOCH = 10
LEAK_RATE = 0.133705 #0.052022
LAMBDA = 0.000000001
```

```python
random.seed(20210805)
np.random.seed(20210805)
result=[]
for LEAK_RATE in tqdm(np.arange(0,1,0.00001)):
    model = SimpleCycleReservoir(train_inputs, train_target, 
                                 NUM_RESERVOIR_LAYER, NUM_INPUT_NODES,
                                 NUM_RESERVOIR_NODES, NUM_OUTPUT_NODES,
                                 LEAK_RATE, NUM_EPOCH,lambda0 = LAMBDA)
    model.train() 
    train_result = model.get_train_result()
    rmse = np.sqrt(MSE(train_inputs[1:],train_result[:-1]))
    result.append(rmse)
    
late_result = pd.DataFrame(result)
late_result  = late_result.set_index(np.arange(0,1,0.00001))
late_result = late_result.rename(columns = {0:'RMSE'}) 
```

```python
late_result.sort_values('RMSE')
```
