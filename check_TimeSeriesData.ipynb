```python
import os
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = open('Henon_Map_time_series_ex1.tex')
data_array=data.readlines()
```

```python
data_array
```

```python
time_data = pd.DataFrame(data_array)
print(type(time_data.iloc[0]))
print(len(time_data))
time_data.head()
```

```python
time_data = time_data.drop(time_data.index[[np.arange(0,len(time_data),2)]])
time_data = time_data.reset_index(drop=True)
```

```python
time_data
```

```python
time_data[0].iloc[0:10] = time_data[0].iloc[0:9].apply(lambda x: x[3:-2])
time_data[0].iloc[10:100] = time_data[0].iloc[10:99].apply(lambda x: x[4:-2])
time_data[0].iloc[100:1000] = time_data[0].iloc[100:999].apply(lambda x: x[5:-2])
time_data[0].iloc[1000:10000] = time_data[0].iloc[1000:10000].apply(lambda x: x[6:-2])
```

```python
time_data[0] = time_data[0].apply(lambda x: float(x))
time_data = time_data.rename(columns={0:'values'})
```

```python
time_data = time_data.dropna().reset_index(drop=True)
```

```python
plt.plot(time_data)
```

```python
time_data.to_csv(os.path.join('Henon_Map_time_series_ex1.csv'))
```
